```python
#Check whether all the zones generate the same amount of sales.
```

```python
# Null Hypothesis :There is no variance across different zones A, B, C and D
# Alternate Hypothesis : NOT all means are equal
```

```python
gp1 = df['Zone - A']
gp2 = df['Zone - B']
gp3 = df['Zone - C']
gp4 = df['Zone - D']
from scipy.stats import f_oneway

f_oneway(gp1, gp2, gp3, gp4)
```

```python
#Since the p-value 0.0011 is lesser than the value of alpha (0.05), we reject the null hyptothesis
#Hence sales of different zones are not same
```
